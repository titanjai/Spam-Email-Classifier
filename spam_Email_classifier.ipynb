```python
import pandas as pd
import string

df = pd.read_csv("spam.csv", encoding="latin1")

df = df[['v1','v2']]
df.columns = ['label','message']

df["label"]=df["label"].map({"ham":0,"spam":1})#label encoding

print(df.head())

print("ORIGINAL:")
print(df['message'].iloc[0])

df['message'] = df['message'].str.lower()#lowercase

print("\nLOWER:")
print(df['message'].iloc[0])

df['message'] = df['message'].apply(
    lambda x: ''.join(char for char in x if char not in string.punctuation)
)#punctuation removal

print("\nNO PUNCTUATION:")
print(df['message'].iloc[0])

tokens = df['message'].iloc[0].split()#tokenization

print("\nTOKENS:")
print(tokens[:10])

print(type(tokens))
```

```python
df["tokens"]=df["message"].apply(lambda x: x.split())
print(df["tokens"].iloc[0])
print(type(df["tokens"].iloc[0]))
```

```python

import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words=set(stopwords.words("english"))
df["tokens"]=df["tokens"].apply(
    lambda x:[word for word in x if word not in stop_words]
)#stopword removal
print(df["tokens"].iloc[0])
```

```python
df["clean_message"]=df["tokens"].apply(
    lambda x:" ".join(x)
)
print(df["clean_message"].iloc[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer#TF-IDF
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(df["clean_message"])
print(x.shape)
```

```python
from sklearn.model_selection import train_test_split#train test split
y=df["label"]
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression#model logistic regression
model=LogisticRegression()
model.fit(X_train,Y_train)
print("training completed")
```

```python
from sklearn.metrics import accuracy_score #accuracy metrics
y_pred=model.predict(X_test)
accuracy=accuracy_score(Y_test,y_pred)
print("accuracy",accuracy)
```

```python
message=["Free entry in a weekly competition . Win cash now!"]
message_tfidf=tfidf.transform(message)
prediction=model.predict(message_tfidf)
print(prediction)
```
